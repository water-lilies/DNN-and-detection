```python
import numpy as np 
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import pandas as pd

from keras.models import Sequential
from keras.layers.core import Dense, Dropout, Activation
from keras.optimizers import SGD
from keras.optimizers import Adam
from keras.callbacks import LambdaCallback
from keras.callbacks import EarlyStopping
from keras.layers import LeakyReLU


def lambdaf_(epoch, logs, step) : 
    if epoch % step == 0 : print(f"{epoch} => {logs}") # 나머지가 0이면 출력           
        
def printepoch(step) :   
    return LambdaCallback(on_epoch_end=lambda epoch, logs: lambdaf_(epoch, logs, step)  )

```

```python
import numpy as np
import numpy.linalg as lin
```

```python
X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([[0],[1],[1],[0]])
```

```python
model = Sequential()
              #입력층 10*2+10 = 30개(학습파라미터수)
model.add(Dense(10, input_dim=2))  # 첫번째 층은 input dimension이 필요
model.add(Activation('sigmoid'))

# 은닉층(hidden layer)
model.add(Dense(10)) # 첫번째 층 이후는 이전층의 output을 input으로
model.add(Activation('sigmoid'))

model.add(Dense(10))
model.add(Activation('sigmoid'))

model.add(Dense(10))
model.add(Activation('sigmoid'))

model.add(Dense(10))
model.add(Activation('sigmoid'))

model.add(Dense(5))
model.add(Activation('sigmoid'))

# 출력층
model.add(Dense(1))
model.add(Activation('sigmoid'))

model.summary() #층이 깊어지면 위로 갈 수록 학습이 덜된다.
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam')
```

```python
model.fit(X, y, batch_size=4, epochs=3000, verbose=0)
print(model.predict_proba(X))  # 행렬은 4x1이다. sample수 x 클래스수 
# 0과 1에 가까운 값이 나와야 함
```

```python
p = model.predict(X)
print((p > 0.5)*1)  #binary로 표현
```

```python
from sklearn.datasets import make_moons
x_data, y_data = make_moons(n_samples=500, noise=0.15)   #  x 는 2차원   y는 레이블    0   1

colors = ["red"  if i == 1  else   "blue"    for i  in y_data  ]

plt.scatter(x_data[:,0], x_data[:,1], color = colors, alpha=1.0)  
print(y_data.shape)
```

```python
model = Sequential([
    Dense(5, activation='sigmoid',input_dim=2),    
    Dense(1, activation='sigmoid'),
])
model.compile(loss='binary_crossentropy', optimizer=Adam(0.01))  
model.fit(x_data, y_data, batch_size=100, epochs=3000, verbose=0,
          validation_data=(x_data, y_data))

colors = ["red"  if i == 1  else   "blue"    for i  in y_data  ]
plt.scatter(x_data[:,0], x_data[:,1], color = colors, alpha=1.0)

xx, yy = np.mgrid[-1.5:2.5:0.1, -1:1.5:0.1]
grid = np.c_[xx.ravel(), yy.ravel()]
h = model.predict_proba(grid)
colors = ["red"  if i>  0.5  else  "blue"  for i  in h  ]
plt.scatter(xx.flatten(), yy.flatten(), color = colors, alpha=0.2)  
```

```python
model = Sequential([
    Dense(5, activation='sigmoid',input_dim=2),   
    Dense(5, activation='sigmoid'),
    Dense(10, activation='sigmoid'),
    Dense(10, activation='sigmoid'),
    Dense(10, activation='sigmoid'),
    Dense(10, activation='sigmoid'),
    Dense(10, activation='sigmoid'),
    Dense(10, activation='sigmoid'),
    Dense(10, activation='sigmoid'),
    Dense(10, activation='sigmoid'),
    Dense(1, activation='sigmoid'),
])
model.compile(loss='binary_crossentropy', optimizer="adam", metrics=['accuracy'])  
```

```python
model.fit(x_data, y_data, batch_size=100, epochs=3000, verbose=0,
          validation_data=(x_data, y_data),
          callbacks=[printepoch(500)])
                     # 500개마다 출력
#loss값이 줄어들지 않는다면, 학습이 안되고 있는 것
```

```python
colors = ["red"  if i == 1  else   "blue"    for i  in y_data  ]
plt.scatter(x_data[:,0], x_data[:,1], color = colors, alpha=1.0)

xx, yy = np.mgrid[-1.5:2.5:0.05, -1:1.5:0.05]
grid = np.c_[xx.ravel(), yy.ravel()]
h = model.predict_proba(grid)
colors = ["red"  if i>  0.5  else  "blue"  for i  in h  ]
plt.scatter(xx.flatten(), yy.flatten(), color = colors, alpha=0.2)  
```

## LeakyReLU
시그모이드는 오래걸리므로 ReLU함수 사용.
<br> 학습이 안되는 경우가 거의 없고 학습 속도가 빠르다.
- LeakyReLU는 발산함수. (ReLU함수는 음수일 때 0 값을 갖는다.)
- 시그모이드는 수렴함수.

참고) 미분값이 0이면 학습이 안된다. w=w-(α*편미분값= 0)이면 w=w이 되므로. 
<br> *α = 학습률

```python
model = Sequential([
    Dense(5, activation=LeakyReLU(alpha=0.1),input_dim=2),   
    Dense(10, activation=LeakyReLU(alpha=0.1)),   # 음수일때의 기울기를 0.1로.
    Dense(10, activation=LeakyReLU(alpha=0.1)),
    Dense(10, activation=LeakyReLU(alpha=0.1)),
    Dense(10, activation=LeakyReLU(alpha=0.1)),
    Dense(10, activation=LeakyReLU(alpha=0.1)),
    Dense(10, activation=LeakyReLU(alpha=0.1)),
    Dense(10, activation=LeakyReLU(alpha=0.1)),
    Dense(1, activation='sigmoid'),   # 마지막 함수는 시그모이드 함수로 써줘야함
])
model.compile(loss='binary_crossentropy', optimizer="adam", metrics=['accuracy']) 
```

```python
model.fit(x_data, y_data, batch_size=100, epochs=3000, verbose=0,
          validation_data=(x_data, y_data),
          callbacks=[printepoch(500)])  
# 빠르게 수렴하는 것을 볼 수 있다. 99%를 보인다.
```

```python
colors = ["red"  if i == 1  else   "blue"    for i  in y_data  ]
plt.scatter(x_data[:,0], x_data[:,1], color = colors, alpha=1.0)

xx, yy = np.mgrid[-1.5:2.5:0.05, -1:1.5:0.05]
grid = np.c_[xx.ravel(), yy.ravel()]
h = model.predict_proba(grid)
colors = ["red"  if i>  0.5  else  "blue"  for i  in h  ]
plt.scatter(xx.flatten(), yy.flatten(), color = colors, alpha=0.2)  
```
